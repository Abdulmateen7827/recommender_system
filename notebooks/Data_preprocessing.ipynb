# User component computed by per genre average rating

```python


# Define the list of genres
genres = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Horror', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'Film-Noir', 'Musical', 'War', 'Western']

# Create a dictionary comprehension to calculate ratings for each genre
genre_ratings = {genre: [df_ratings['rating'][i] if genre in item_train_df['genres'][i].split('|') else 0 for i in range(len(item_train_df['movieId']))] for genre in genres}

# Convert the dictionary to a DataFrame
user_train_df = pd.DataFrame(genre_ratings)


```

```python
user_train_df
```

```python
user_train_df = pd.concat([df_ratings['userId'],user_train_df],axis=1)
```

```python

# Define the list of genre columns
genre_columns = ['Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Horror', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'Film-Noir', 'Musical', 'War', 'Western']

# Iterate over each user ID
for i in range(1, 611):
    # Calculate the average ratings for the genres
    average_ratings = user_train_df.loc[user_train_df['userId'] == i, genre_columns]
    average_ratings = round(average_ratings[average_ratings != 0].mean(), 1)

    # Update the genre columns with the respective average ratings
    for genre in genre_columns:
        user_train_df.loc[user_train_df['userId'] == i, genre] = average_ratings[genre]


```

```python
user_train_df =  user_train_df.fillna(0)
```

```python
user_train_df.to_csv('content_user_train',index=False,header=False)
```
